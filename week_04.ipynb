## Frequency Distribution
- 단어의 빈도수를 활용할 수 있는 많은 메소드 제공
- 자연어처리: 인간의 언어 -> 컴퓨터가 인식할 수 있게 하는 것
    - 전체 단어 수
    - 특정 단어의 빈도수

```python
fdist = FreqDist(text1)
fdist 
# fdist = FreqDist(text) : text내의 단어들에 대해 빈도 분포표(frequency distribution)를 만든다.
# FreqDist라는 클래스에 text1을 넣으면 실행됨
# dictinary 타입으로 반환됨 | key:value(나중에 배움)
# 문서에 사용된 단어(토큰)의 빈도수를 알려주는 클래스
```

```python
fdist.most_common(10)
# 가장 출현 횟수가 높은 단어를 찾음
# fdist에 FreqDist(text1)을 저장해서 쓰는거임
```

```python
fdist['whale'], text1.count('whale'), FreqDist(text1)['whale']
# 특정 단어 key에 접근해서 value인 빈도수를 출력. 즉, 특정 단어의 빈도수 출력
# 3개 동일
```

```python
fdist.freq('whale')
# 전체에서 해당 단어가 차지하는 비율
```

```python
fdist.plot(50, cumulative=True)
# 50위까지 누적 분포 그래프를 그린다.
```

```python
fdist.plot(50, cumulative=False)
```

```python
# 1번 나타나는 단어들
fdist.hapaxes()
```

```python
# 1번 나타나는 단어들의 갯수, 중복 제외, 대소문자 구분됨
len(fdist.hapaxes())
```

## list comprehension

- 조건을 만족시키는 단어만 떼어오자
- 조건+ : [A for A' in if <조건식> and <조건식>...]

## 조건을 위한 함수 : uncased, cased
- 'the' != 'The'
- 형식: string.메소드()
- upper(), lower(), title()
    - upper() : 문자열을 모두 대문자로 변경
    - lower() : 문자열을 모두 소문자로 변경
    - title() : 주어진 문자열에서 알파벳 외의 문자(숫자, 특수기호, 띄어쓰기 등)로 나누어져 있는 영단어들의 첫 글자를 모두 대문자로 변환시킨다.
- capitalize(), swapcase()
    - capitalize() : 주어진 문자열에서 맨 첫 글자를 대문자로 변환시킨다.
    - swapcase() : 대문자는 소문자로, 소문자는 대문자로
- isupper(), islower(), istitle()
    - isupper() : 문자가 대문자인지 확인
    - islower() : 소문자인지 확인
    - istitle() : 주어진 문자열에서 나누어져있는 영단어의 첫글자가 대문자인지 검사
    - 단, 'GOD'.istitle()는 false임. 즉, 모두 대문자면 안됨
- endswith(), startswith(), isalpha()
    - endswith(str) : 특정 문자열로 끝나는지 검사(T/F)
    - startswith(str) : 특정 문자열로 시작하는지 검사(T/F)
    - isalpha() : 문자열이 알파벳으로만 이루어져있는지 검사

```python
# 정답: 리스트 컴프리헨션 이용

sent = 'I, am, thirsty'
tokens = sent.split(',') # ',' 기준으로 잘라서 리스트에 저장, ['I', ' am', ' thirsty']

tokens = [word.strip() for word in tokens] # => ['I', 'am', 'thirsty']
print(tokens)
```

```python
name = 'son'
age = 22
print('My name is', name)
```

```python
print(f'My name is {name} and I am {age} years old') # f스트링
```
